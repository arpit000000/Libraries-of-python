```python
sts=pd.DataFrame({'Name':['arpit','anuj','ram']
                 ,'Teams':['alpha','beta','gamma']})
sts
```

```python
st=pd.DataFrame({
    'Name':['arpit','ram','anuj','raj'],'marks':[30,12,24,87]
})
st
```

```python
pd.merge(sts,st)
```

```python
rst=pd.merge(sts,st,on='Name',how='outer')
rst
```

```python
rst=pd.merge(sts,st,on='Name',how='inner')
rst
```

```python
rst=pd.merge(sts,st,on='Name',how='left')
rst
```

```python
rst=pd.merge(sts,st,on='Name',how='right')
rst
```

```python
rst=pd.merge(sts,st,on='Name',how='outer',indicator=True)
rst
```

```python
sts['marks']=[80,90,100]
```

```python
rst=pd.merge(sts,st,on='Name',how='outer',suffixes=('_sts','_st'))
rst
```
